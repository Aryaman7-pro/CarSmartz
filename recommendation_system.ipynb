```python
categorical_features = ['fuel_type', 'company']
numerical_features = ['year', 'predicted_price', 'kilo_driven']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])
```

```python
car_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('knn', NearestNeighbors(n_neighbors=5, algorithm='auto'))  # Adjust n_neighbors as needed
])
```

```python
df1 = pd.read_csv('C:\\Users\\aryaman.kanwar\\CarSmartz\\cmz_car.csv')
```

```python
df1[0:20]
```

```python
car_pipeline.fit(df1)
```

```python
def recommend_similar_cars(car_model):
    # Check if the car model exists in the DataFrame
    if car_model not in df1['car_model'].values:
        print(f"Car model '{car_model}' not found in the dataset.")
        return pd.DataFrame()  # Return an empty DataFrame if not found
    
    # Get the index of the specified car model
    car_index = df1[df1['car_model'] == car_model].index[0]
    
    # Transform the features of the selected car using the preprocessor
    car_features = car_pipeline.named_steps['preprocessor'].transform(df1.iloc[[car_index]])
    
    # Get similar cars based on the selected car's index
    distances, indices = car_pipeline.named_steps['knn'].kneighbors(car_features)
    
    # Extract the similar cars and drop unnecessary columns
    similar_cars = df1.iloc[indices[0]].drop(columns=['year', 'predicted_price', 'kilo_driven', 'fuel_type', 'company']).reset_index(drop=True)
    
    return similar_cars
```

```python
recommended_cars = recommend_similar_cars('Honda Amaze 1.2')
print(recommended_cars)
```
